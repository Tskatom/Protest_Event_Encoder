```python
pop_file = "../../data/single_label/pop_labels"
type_file = "../../data/single_label/type_labels"

pops = [l.strip() for l in open(pop_file)]
types = [l.strip() for l in open(type_file)]
```

```python
import pandas as pds
class_names = {"011": "Employment and Wages", "012":"Housing", "013": "Energy and Resources", "014": "Economic Policy",
               "015": "Government Policy", "016": "Others"}
com_dic = {}
for p, t in zip(pops, types):
    if p not in com_dic:
        com_dic[p] = {}
    t = class_names[t]
    if t not in com_dic[p]:
        com_dic[p][t] = 0
    com_dic[p][t] += 1
df = pds.DataFrame(com_dic)
```
